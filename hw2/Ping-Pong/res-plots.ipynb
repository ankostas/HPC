```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('res-2.txt', delimiter = '\t')
data = pd.DataFrame(data, columns=["Size(Bytes)", "Iterations","Total time(sec)", "Time per message(sec)","Bandwidth(Bytes/s)"])
display(data)

fig, ax = plt.subplots()

ax.plot(data["Size(Bytes)"], data["Time per message(sec)"], 'b')
ax.set_xlabel("Size(Bytes)")
ax.set_ylabel("Time per message(sec)")

ax2 = ax.twinx()

ax2.plot(data["Size(Bytes)"], data["Bandwidth(Bytes/s)"], 'r')
ax2.set_xlabel("Size(Bytes)")
ax2.set_ylabel("Bandwidth(Bytes/s)")
plt.grid()
plt.show()
```
