```python
import numpy as np
from IPython import display
import matplotlib.pyplot as plt
import time
```

```python
data = np.loadtxt("res.txt")
N = np.sqrt(data.shape[1]).astype('int')
N
```

```python
for vals in data:
    display.clear_output(wait=True)
    plt.imshow(vals.reshape((N, N)))
    plt.show()
    time.sleep(1e-1)
```
