```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

filename = "res-"
N = str(100)
m=8
plt.figure(figsize=(15,10))
for i in range(1, 7):
    res = np.loadtxt(filename + str(i), delimiter='\t')
    plt.plot(res[:, 0], res[0, 1]/res[:, 1], '.-', label="N:{}".format(N))
    N = N + str(0)
plt.xlabel("Number of processes")
plt.ylabel("Speedup")
plt.legend()
plt.grid()
plt.show()
```
